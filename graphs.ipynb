```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
x=["apples","bananas","cherries","dates"]
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["apples","bananas","cherries","dates"]
myexplode=[0.2,0,0,0.5]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```
